```python
# Gemini API Key test with LangChain
import os
from dotenv import load_dotenv
from langchain_google_genai import ChatGoogleGenerativeAI

# Load environment variables from .env
load_dotenv()

from langchain.agents import create_agent

def getweather(city: str) -> str:
    """Get weather for a given city (e.g., San Francisco, Delhi)"""
    return f"The weather in {city} is "

llm = ChatGoogleGenerativeAI(
    google_api_key=os.getenv('gemini_api_key'),
    model="gemini-3.1-flash-lite-preview"
)
agent = create_agent(
    model=llm,
    tools=[getweather],
    system_prompt="you are an obedient ai assista nt"
)

result = agent.invoke({
    "messages": [
        {"role": "user", "content": "What is the weather in San Francisco?"}
    ]
})
# This gets the last message in the list (the AI's reply) and extracts its text
print(result["messages"][-1].content)
```

### Calling the gemini llm using langchain's init caht model 

```python
import os
from dotenv import load_dotenv
from langchain.chat_models import init_chat_model

# 1. Load the .env file
load_dotenv()

# 2. Map your key to the standard Google name
# (Only needed if your .env variable isn't already named GOOGLE_API_KEY)
os.environ["GOOGLE_API_KEY"] = os.getenv('gemini_api_key')

# 3. Initialize the model
llm = init_chat_model(
    "gemini-3.1-flash-lite-preview", 
    model_provider="google_genai",
    temperature=0 
)

# 4. Run it
try:
    response = llm.invoke("Why do parrots talk")
    print(response.content)
except Exception as e:
    print(f"Error: {e}")
```
