### simple gemini ai api testing ###

```python
# Gemini API Key test with LangChain
import os
from dotenv import load_dotenv
from langchain_google_genai import ChatGoogleGenerativeAI

# Load environment variables from .env
load_dotenv()

# Get Gemini API key
api_key = os.getenv('gemini_api_key')
assert api_key, "Gemini API key not found in .env file!"

# Create Gemini chat model
llm = ChatGoogleGenerativeAI(
    google_api_key=api_key,
    model="gemini-3.1-flash-lite-preview"
)

# Test prompt
response = llm.invoke("Hello Gemini! Can you tell me a fun fact?")
print(response.content)
```

### BUilding an agent with gemini api key 

```python
# Gemini API Key test with LangChain
import os
from dotenv import load_dotenv
from langchain_google_genai import ChatGoogleGenerativeAI

# Load environment variables from .env
load_dotenv()

from langchain.agents import create_agent

def getweather(city: str) -> str:
    """Get weather for a given city (e.g., San Francisco, Delhi)"""
    return f"The weather in {city} is "

llm = ChatGoogleGenerativeAI(
    google_api_key=os.getenv('gemini_api_key'),
    model="gemini-3.1-flash-lite-preview"
)
agent = create_agent(
    model=llm,
    tools=[getweather],
    system_prompt="you are an obedient ai assistant"
)

result = agent.invoke({
    "messages": [
        {"role": "user", "content": "What is the weather in San Francisco?"}
    ]
})
# This gets the last message in the list (the AI's reply) and extracts its text
print(result["messages"][-1].content)
```

### Simple groq api key testing 

```python
# Groq API Key test with LangChain
import os
from dotenv import load_dotenv
from langchain_groq import ChatGroq

# Load environment variables from .env
load_dotenv()

# Get Groq API key
api_key = os.getenv('groq_api_key')
assert api_key, "Groq API key not found in .env file!"

# Create Groq chat model
llm = ChatGroq(
    groq_api_key=api_key,
    model="groq/compound"
)

# Test prompt
response = llm.invoke("Hello Groq! Can you tell me a fun fact?")
print(response.content)
```
